QSN-1-What is hypothesis testing in statistics.
ANS-
Hypothesis testing is a statistical technique used to make inferences about a population based on a sample of data. It involves testing a hypothesis about the population parameter, such as the mean, proportion, or variance, to determine whether it is true or false.

Common hypothesis tests:

- t-tests: Compare means of two groups
- ANOVA: Compare means of three or more groups
- Chi-squared tests: Compare proportions of two or more groups
- Regression analysis: Model the relationship between a dependent variable and one or more independent variables.


QSN-2-What is the null hypothesis, and how does it differ from the alternative hypothesis.
ANS-
In statistical hypothesis testing, the null hypothesis (H0) and alternative hypothesis (H1) are two mutually exclusive statements that are tested against each other.

Null Hypothesis (H0):

The null hypothesis is a statement of no effect, no difference, or no relationship. It is a statement that there is no significant difference between the variables being tested. The null hypothesis is often denoted as H0.

In summary, the null hypothesis is a statement of no effect or no difference, while the alternative hypothesis is a statement that contradicts the null hypothesis. The null hypothesis is assumed to be true unless there is sufficient evidence to reject it.

The P-value tells us the probability of obtaining the observed result (or a more extreme result) if the null hypothesis is true. It does not tell us the probability that the null hypothesis is true.

QNS-6-What are Type 1 and Type 2 errors in hypothesis testing
ANS-

Type I Error:
A Type I error occurs when we reject the null hypothesis when it is actually true. This is also known as a "false positive" error.

Example:
Suppose we conduct a hypothesis test to determine whether a new medication is effective in reducing blood pressure. We reject the null hypothesis and conclude that the medication is effective, but in reality, it has no effect. This is a Type I error.

Type II Error:
A Type II error occurs when we fail to reject the null hypothesis when it is actually false. This is also known as a "false negative" error.

Example:
Suppose we conduct a hypothesis test to determine whether a new medication is effective in reducing blood pressure. We fail to reject the null hypothesis and conclude that the medication has no effect, but in reality, it is effective. This is a Type II error.

QSN-15-What is the margin of error, and how does it affect the confidence interval.

The margin of error (ME) is a statistical measure that represents the maximum amount by which the sample estimate may differ from the true population parameter. It is a critical component of confidence intervals, as it determines the width of the interval.
Factors Affecting the Margin of Error:
1. Sample size (n): Increasing the sample size reduces the margin of error.
2. Confidence level: Increasing the confidence level (e.g., from 90% to 95%) increases the margin of error.
3. Standard deviation (σ): Increasing the standard deviation increases the margin of error.

Calculating the Margin of Error:
The margin of error can be calculated using the following formula:

ME = (Z * σ) / √n
Where:

- ME = margin of error
- Z = Z-score corresponding to the desired confidence level
- σ = standard deviation of the population
- n = sample size

Effect of Margin of Error on Confidence Interval:
1. Width of the interval: The margin of error determines the width of the confidence interval. A larger margin of error results in a wider interval.
2. Precision of the estimate: A smaller margin of error indicates a more precise estimate of the population parameter.

```python
##QSN-1-Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
##ANS-
import numpy as np
from scipy import stats
def z_test(sample_mean, population_mean, sample_std, sample_size):

    # Calculate the standard error
    std_error = sample_std / np.sqrt(sample_size)

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / std_error

    # Calculate the P-value
    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))

    return z_score, p_value

# Example usage
sample_mean = 25.5
population_mean = 25.0
sample_std = 2.5
sample_size = 100

z_score, p_value = z_test(sample_mean, population_mean, sample_std, sample_size)

print("Z-score:", z_score)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")

```

```python
#QSN-3-Implement a one-sample Z-test using Python to compare the sample mean with the population mean.
#ANS-
def one_sample_z_test(sample_data, population_mean, population_std):
    """
    Perform a one-sample Z-test to compare the sample mean with the population mean.

    Parameters:
    sample_data (array-like): Sample data
    population_mean (float): Known population mean
    population_std (float): Known population standard deviation

    Returns:
    z_score (float): Z-score
    p_value (float): P-value
    """
    # Calculate the sample mean
    sample_mean = np.mean(sample_data)

    # Calculate the standard error
    std_error = population_std / np.sqrt(len(sample_data))

    # Calculate the Z-score
    z_score = (sample_mean - population_mean) / std_error

    # Calculate the P-value
    p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))

    return z_score, p_value
# Example usage
np.random.seed(0)
sample_data = np.random.normal(10, 2, 100)  # sample data with mean 10 and std 2
population_mean = 9  # known population mean
population_std = 2  # known population standard deviation

z_score, p_value = one_sample_z_test(sample_data, population_mean, population_std)

print("Z-score:", z_score)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

```python
#QSN-5-Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
#ANS-
def calculate_errors(alpha, beta, mu0, mu1, sigma, num_simulations):
    # Initialize error counters
    type1_errors = 0
    type2_errors = 0

    # Perform simulations
    for _ in range(num_simulations):
        # Generate sample data under the null hypothesis
        sample_null = np.random.normal(mu0, sigma, 100)

        # Generate sample data under the alternative hypothesis
        sample_alt = np.random.normal(mu1, sigma, 100)

        # Calculate the test statistic under the null hypothesis
        t_stat_null = np.abs(np.mean(sample_null) - mu0) / (sigma / np.sqrt(100))

        # Calculate the test statistic under the alternative hypothesis
        t_stat_alt = np.abs(np.mean(sample_alt) - mu0) / (sigma / np.sqrt(100))
        # Check for Type 1 error
        if t_stat_null > stats.t.ppf(1 - alpha / 2, 99):
            type1_errors += 1

        # Check for Type 2 error
        if t_stat_alt < stats.t.ppf(1 - alpha / 2, 99):
            type2_errors += 1

    # Calculate error rates
    type1_error_rate = type1_errors / num_simulations
    type2_error_rate = type2_errors / num_simulations

    return type1_error_rate, type2_error_rate

# Define parameters
alpha = 0.05  # significance level
beta = 0.2  # probability of Type 2 error
mu0 = 0  # population mean under the null hypothesis
mu1 = 1  # population mean under the alternative hypothesis
sigma = 1  # population standard deviation
num_simulations = 10000  # number of simulations

# Calculate error rates
type1_error_rate, type2_error_rate = calculate_errors(alpha, beta, mu0, mu1, sigma, num_simulations)

print("Type 1 error rate:", type1_error_rate)
print("Type 2 error rate:", type2_error_rate)
# Visualize error rates
plt.bar(["Type 1 Error", "Type 2 Error"], [type1_error_rate, type2_error_rate])
plt.xlabel("Error Type")
plt.ylabel("Error Rate")
plt.title("Error Rates in Hypothesis Testing")
plt.show()
```

```python
#QSN-9-Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
#ANS-
def calculate_confidence_interval(sample_data, confidence_level=0.95):
# Calculate the sample mean and standard deviation
    sample_mean = np.mean(sample_data)
    sample_std = np.std(sample_data, ddof=1)

    # Calculate the standard error
    std_error = sample_std / np.sqrt(len(sample_data))

    # Determine the degrees of freedom
    df = len(sample_data) - 1

    # Calculate the critical value using the T-distribution
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df)

    # Calculate the margin of error
    margin_of_error = critical_value * std_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound
# Example usage
np.random.seed(0)
sample_data = np.random.normal(10, 2, 100)  # sample data with mean 10 and std 2

confidence_level = 0.95
lower_bound, upper_bound = calculate_confidence_interval(sample_data, confidence_level)

print(f"Confidence Interval ({confidence_level*100}%): ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#QSN-10-Write a Python program to calculate the margin of error for a given confidence level using sample data
#ANS-
def calculate_margin_of_error(sample_data, confidence_level):
   # Calculate the sample size
    sample_size = len(sample_data)

    # Calculate the sample standard deviation
    sample_std = np.std(sample_data, ddof=1)

    # Determine the degrees of freedom
    df = sample_size - 1

    # Calculate the critical value using the T-distribution
    critical_value = stats.t.ppf((1 + confidence_level) / 2, df)

    # Calculate the margin of error
    margin_of_error = critical_value * (sample_std / np.sqrt(sample_size))

    return margin_of_error

# Example usage
np.random.seed(0)
sample_data = np.random.normal(10, 2, 100)  # sample data with mean 10 and std 2

confidence_level = 0.95
margin_of_error = calculate_margin_of_error(sample_data, confidence_level)

print(f"Margin of Error ({confidence_level*100}%): {margin_of_error:.2f}")

# Interpret the results
print(f"There is a {confidence_level*100}% chance that the true population mean lies within {margin_of_error:.2f} units of the sample mean.")
```

```python
#QSN-19-Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA
#ANS-


def check_anova_assumptions(data, group_column, value_column):
    # Check normality assumption
    print("Normality Test:")
    for group in data[group_column].unique():
        group_data = data.loc[data[group_column] == group, value_column]
        w_stat, p_val = stats.shapiro(group_data)
        print(f"Group {group}: W-statistic = {w_stat:.4f}, P-value = {p_val:.4f}")
        if p_val < 0.05:
            print(f"Group {group} does not meet the normality assumption.")
        else:
            print(f"Group {group} meets the normality assumption.")

    # Check independence assumption
    print("\nIndependence Test:")
    corr_coef = data.groupby(group_column)[value_column].corrwith(data[group_column])
    print(f"Correlation Coefficient: {corr_coef:.4f}")
    if np.abs(corr_coef) < 0.1:
        print("Independence assumption is met.")
    else:
        print("Independence assumption is not met.")

    # Check equal variance assumption
    print("\nEqual Variance Test:")
    for group in data[group_column].unique():
        group_data = data.loc[data[group_column] == group, value_column]
        variance = np.var(group_data, ddof=1)
        print(f"Group {group}: Variance = {variance:.4f}")
    levene_stat, p_val = stats.levene(*[data.loc[data[group_column] == group, value_column] for group in data[group_column].unique()])
    print(f"Levene's Test: W-statistic = {levene_stat:.4f}, P-value = {p_val:.4f}")
    if p_val < 0.05:
        print("Equal variance assumption is not met.")
    else:
        print("Equal variance assumption is met.")

    # Visualize data using boxplot and Q-Q plot
    plt.figure(figsize=(12, 6))

    plt.subplot(1, 2, 1)
    plt.boxplot([data.loc[data[group_column] == group, value_column] for group in data[group_column].unique()], labels=data[group_column].unique())
    plt.xlabel("Group")
    plt.ylabel("Value")
    plt.title("Boxplot")

    plt.subplot(1, 2, 2)
    stats.probplot(data[value_column], dist="norm", plot=plt)
    plt.title("Q-Q Plot")

    plt.tight_layout()
    plt.show()

# Example usage
np.random.seed(0)
data = pd.DataFrame({
    "Group": np.repeat(["A", "B", "C"], 10),
    "Value": np.random.normal(0, 1, 30)
})
check_anova_assumptions(data, "Group", "Value")
```
